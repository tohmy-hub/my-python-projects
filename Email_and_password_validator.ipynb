```python
import re
print("EMAIL VALIDATOR")

def validate_email(email: str):
  errors = []

  if len(email) < 8:
    errors.append("email must be at least 8 character long.")
  if "@" not in email:
    errors.append("Email must contain this special character(@)")
  if ".com" not in email:
    errors.append("Email must end with (e.g. .com,.yahoo)")

  return (len(errors) == 0), errors

email = input("Enter your email: ")

is_valid, problems = validate_email(email)

if is_valid:
    print("✅ Email is strong!")
else:
    print("❌ Email is weak. Issues:")
    for error in problems:
        print("-", error)

import re

def validate_password(password):
    errors = []
    score = 0  # We'll use this to rate strength

    # Check length
    if len(password) >= 8:
        score += 1
    else:
        errors.append("Password must be at least 8 characters long.")

    # Uppercase letter
    if re.search(r"[A-Z]", password):
        score += 1
    else:
        errors.append("Password must contain at least one uppercase letter.")

    # Lowercase letter
    if re.search(r"[a-z]", password):
        score += 1
    else:
        errors.append("Password must contain at least one lowercase letter.")

    # Number
    if re.search(r"\d", password):
        score += 1
    else:
        errors.append("Password must contain at least one number.")

    # Special character
    if re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):
        score += 1
    else:
        errors.append("Password must contain at least one special character.")

    # Determine strength
    if score <= 2:
        strength = "Weak 😞"
    elif score == 3 or score == 4:
        strength = "Medium 😐"
    else:
        strength = "Strong 💪"

    return strength, errors


# === Main program ===
password = input("Enter your password: ")

strength, problems = validate_password(password)

print(f"\nPassword Strength: {strength}")

if problems:
    print("Issues:")
    for error in problems:
        print("-", error)
else:
    print("✅ Your password meets all requirements!")
```
